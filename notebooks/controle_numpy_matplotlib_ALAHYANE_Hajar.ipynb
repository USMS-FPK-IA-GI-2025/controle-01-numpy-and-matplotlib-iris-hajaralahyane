# Partie A — Importation du Dataset, Arrays, Premières Manipulations

### A1. Importation du dataset
Importez NumPy et Matplotlib.  
Chargez le fichier `iris.csv` avec `np.genfromtxt` (`delimiter=','`, `skip_header=1`).  
Stockez les données dans un array nommé `data` et affichez sa `shape`.


```python
# Votre code ici
import numpy as np
import matplotlib.pyplot as plt
data = np.genfromtxt('/content/Iris.csv', delimiter=',', skip_header=1)
print(data.shape)
```

### A2. Séparation des features et des espèces
Séparez :
- les 4 premières colonnes numériques dans un array `features`
- la dernière colonne (espèces) dans un array `species`
Utilisez le slicing.


```python
# Votre code ici
#Séparation des features(4 premières colonnes)
features=data[:,:4]

#Séparation des espèces
species=data[:,-1]

#Affichage
print('Les 4 premières colnnes:',features.shape)
print('La dernière colonne:',species.shape)
```

# Partie B — Statistiques Basiques et Opérations Mathématiques

### B6. Moyenne par colonne
Calculez la moyenne de chaque colonne de `features` avec `np.mean(axis=0)`.


```python
# Votre code ici
#Moyenne par colonne
Moyenne=np.mean(features,axis=0)
print(Moyenne)
```

### B7. Min et Max
Trouvez la valeur minimale et maximale de la longueur des pétales.


```python
# Votre code ici
Min=np.min(features[:,3])
Max=np.max(features[:,3])
print('La valeur minimale de pétale :',Min)
print('La valeur maximale de pétale :',Max)
```

### B9. Normalisation
Normalisez les longueurs des pétales en les divisant par leur maximum.
Stockez le résultat dans `normalized_petals`.


```python
# Votre code ici
normalized_petals=features[:,3]/Max
print('Normalisation des longeurs des pétales :',normalized_petals)
```

### B10. Somme cumulée
Calculez la somme cumulée des largeurs des pétales pour Versicolor.
Affichez les 10 premières valeurs.


```python
# Votre code ici
Somme=np.cumsum(features[50:100,3][:10])
print('la somme cumulée pour Versicolor est :',Somme)
```

### C14. Data cleaning (simulation)
Simulez des valeurs NaN et remplacez-les avec `np.nan_to_num`.


```python
# Votre code ici
feautures_nan=features.copy()
feautures_nan[0,0]=np.nan
feautures_nan[10,2]=np.nan
feautures_nan[25,3]=np.nan
print('Les feautures avec nan:',feautures_nan)
feautures_nan_num=np.nan_to_num(feautures_nan)
print('Les feauters remplcés par un nombre :',feautures_nan_num)
```

# Partie D — Visualisation avec Matplotlib (Plots Simples)

### D17. Line plot
Line plot des longueurs des sépales pour les 50 premières fleurs.


```python
# Votre code ici
sepal_length_50=features[:50,0]
plt.figure(figsize=(10,5))
plt.plot(sepal_length_50)
plt.xlabel('Index')
plt.ylabel('Longueur des sépales')
plt.title('Line plot des longueurs des sépales pour les 50 premières fleurs')
plt.show()
```

### D18. Scatter plot
Scatter plot : longueur vs largeur des pétales.


```python
# Votre code ici
petal_lenght=features[:,2]
petal_width=features[:,3]
plt.figure(figsize=(10,5))
plt.scatter(petal_lenght,petal_width,color='red')
plt.xlabel('Longueur des pétales')
plt.ylabel('Largeur des pétales')
plt.title('Scatter plot : longueur vs largeur des pétales')
plt.show()
```

### D19. Bar chart
Bar chart des moyennes des quatre caractéristiques.


```python
# Votre code ici
feature_name=['sepal_length','sepal_width','petal_length','petal_width']
plt.figure(figsize=(10,5))
plt.bar(feature_name,Moyenne,color='hotpink')
plt.xlabel('Features')
plt.ylabel('Moyenne')
plt.title('Bar chart des moyennes des quatre caractéristiques')
plt.show()
```

### D20. Histogram
Histogramme de la largeur des sépales (10 bins).


```python
# Votre code ici
sepal_width=features[:,1]
plt.figure(figsize=(10,5))
plt.hist(sepal_width,bins=10,color='green')
plt.xlabel('Largeur des sépales')
plt.ylabel('Fréquence')
plt.title('Histogramme de la largeur des sépales (10 bins)')
plt.show()
```

### D21. Boxplot par espèce
Boxplot des longueurs des pétales par espèce.


```python
# Votre code ici
petal_lenght_setosa=features[0:50,2]
petal_lenght_versicolor=features[50:100,2]
petal_lenght_virginica=features[100:150,2]
plt.figure(figsize=(10,5))
plt.boxplot([petal_lenght_setosa,petal_lenght_versicolor,petal_lenght_virginica],labels=['Setosa','Versicolor','Virginica'])
plt.xlabel('Espèces')
plt.ylabel('Longueur des pétales')
plt.title('Boxplot des longueurs des pétales par espèce')
plt.show()
```
